```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

dataHW = pd.read_csv('height_weight.csv')

gender = []
for x in dataHW['Gender']:
    if x == 1:
        gender.append('Male')
    if x == 2:
        gender.append('Female')

dataHW['Gender'] = gender
#print(data.head())

plt.scatter(dataHW['Height'],dataHW['Weight'])
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show
sns.relplot(x='Height',y='Weight',hue="Gender",s=200, alpha=.5, data=dataHW)
```
